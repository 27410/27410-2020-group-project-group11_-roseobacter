Drawing the map of our model in Escher

```python
import escher
import cobra.io
model = cobra.io.read_sbml_model('Roseobacter-litoralis-strain-B14.SBML')
```

```python
escher.list_available_maps()
```

```python
escher.Builder(map_name = 'e_coli_core.Core metabolism', model = model, highlight_missing = True).display_in_notebook()
```

```python
cobra.io.save_json_model(model, "carveme_model.json")
```
